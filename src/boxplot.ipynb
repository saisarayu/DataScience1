```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create dataset
data = {
    "math_marks": [60, 65, 70, 75, 80],
    "science_marks": [60, 65, 70, 75, 120],  # outlier
    "english_marks": [68, 69, 70, 71, 72]   # consistent
}

df = pd.DataFrame(data)

# Step 2: Boxplot
plt.boxplot([df["math_marks"], df["science_marks"], df["english_marks"]],
            labels=["Math", "Science", "English"])

plt.title("Boxplot Comparison of Marks")
plt.ylabel("Marks")

plt.show()
```
