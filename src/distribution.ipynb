```python
import pandas as pd

# Step 1: Create dataset (multiple numeric columns)
data = {
    "math_marks": [60, 65, 70, 75, 80],
    "science_marks": [60, 65, 70, 75, 120],  # outlier
    "english_marks": [68, 69, 70, 71, 72]   # stable
}

df = pd.DataFrame(data)

print("Dataset:\n", df)


# Step 2: Summary statistics
print("\nSummary Statistics:")
print(df.describe())

print("\nMedian values:")
print(df.median())


# Step 3: Compare distributions
print("\n--- Comparison ---")

print("Means:")
print(df.mean())

print("\nStandard Deviations:")
print(df.std())


# Step 4: Interpretation (PRINT THIS — most people skip)
print("\n--- Interpretation ---")

print("Science has higher mean due to an outlier (120).")
print("English has low standard deviation → very consistent values.")
print("Science has high standard deviation → values are spread out.")
print("Math is moderate → balanced distribution.")
```
