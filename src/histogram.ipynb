```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create dataset
data = {
    "math_marks": [50, 55, 60, 65, 70, 75, 80],
    "science_marks": [50, 55, 60, 65, 70, 75, 120]  # outlier → skew
}

df = pd.DataFrame(data)

# Step 2: Histogram for Math
plt.hist(df["math_marks"], bins=5)
plt.title("Math Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# Step 3: Histogram for Science
plt.hist(df["science_marks"], bins=5)
plt.title("Science Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```
